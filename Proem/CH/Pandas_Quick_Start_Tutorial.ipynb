#### 2.10 åˆ†ç±»æ•°æ® <a id='2-10'></a>

è¿™ä¸€éƒ¨åˆ†ä»‹ç»äº† Pandas ä¸­çš„åˆ†ç±»æ•°æ®ç±»å‹ï¼ˆCategorical data typeï¼‰ã€‚åˆ†ç±»æ•°æ®ç±»å‹å¯ä»¥ç”¨æ¥è¡¨ç¤ºç±»åˆ«å˜é‡ï¼Œå®ƒåŒ…å«æœ‰é™æ•°é‡çš„å¯èƒ½å€¼ã€‚åœ¨ Pandas ä¸­ï¼Œä½ å¯ä»¥ä½¿ç”¨ pd.Categorical æ¥åˆ›å»ºä¸€ä¸ªåˆ†ç±»å˜é‡ã€‚

````python
df = pd.DataFrame(
        {"id": [1, 2, 3, 4, 5, 6], "raw_grade": ["a", "b", "b", "a", "a", "e"]}
   )
df
````

å°†åŸå§‹æˆç»©è½¬æ¢ä¸ºåˆ†ç±»æ•°æ®ç±»å‹ï¼š

````python
df["grade"] = df["raw_grade"].astype("category")

df["grade"]
````

å°†ç±»åˆ«é‡å‘½åä¸ºæ›´æœ‰æ„ä¹‰çš„åç§°ï¼š

````python
new_categories = ["very good", "good", "very bad"]

df["grade"] = df["grade"].cat.rename_categories(new_categories)
df["grade"]
````

å¯¹ç±»åˆ«é‡æ–°æ’åºå¹¶åŒæ—¶æ·»åŠ ç¼ºå°‘çš„ç±»åˆ«ï¼ˆé»˜è®¤æƒ…å†µä¸‹ï¼ŒSeries.cat() ä¸‹çš„æ–¹æ³•è¿”å›ä¸€ä¸ªæ–°çš„ Seriesï¼‰ï¼š

#### 1.3 éªŒè¯ TensorFlow å®‰è£…å¹¶æŸ¥çœ‹ç‰ˆæœ¬ <a id='1-3'></a>

å®‰è£… TensorFlow åï¼Œç¡®è®¤åº“å·²æˆåŠŸå®‰è£…å¹¶ä¸”ä½ å¯ä»¥å¼€å§‹ä½¿ç”¨å®ƒã€‚ 

ä¸è¦è·³è¿‡æ­¤æ­¥éª¤ã€‚ 

å¦‚æœ TensorFlow æœªæ­£ç¡®å®‰è£…æˆ–åœ¨æ­¤æ­¥éª¤ä¸­å¼•å‘é”™è¯¯ï¼Œåˆ™å°†æ— æ³•è¿è¡Œä¹‹åçš„ç¤ºä¾‹ã€‚

### 2 TensorFlow æ–¹æ³• <a id='tensorflowfeatures'></a>

é¦–å…ˆå°† TensorFlow å¯¼å…¥åˆ°æ‚¨çš„ç¨‹åºï¼š

```python
import tensorflow as tf
print("TensorFlow ç‰ˆæœ¬:", tf.__version__)
```

#### 2.1 åŠ è½½æ•°æ®é›† <a id='2-1'></a>

åŠ è½½å¹¶å‡†å¤‡ [MNIST æ•°æ®é›†](http://yann.lecun.com/exdb/mnist/)ã€‚å°†æ ·æœ¬æ•°æ®ä»æ•´æ•°è½¬æ¢ä¸ºæµ®ç‚¹æ•°ï¼š

```python
mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

#### 2.2 æ„å»ºæœºå™¨å­¦ä¹ æ¨¡å‹ <a id='2-2'></a>

é€šè¿‡å †å å±‚æ¥æ„å»º `tf.keras.Sequential` æ¨¡å‹ã€‚

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10)
])
```

å¯¹äºæ¯ä¸ªæ ·æœ¬ï¼Œæ¨¡å‹éƒ½ä¼šè¿”å›ä¸€ä¸ªåŒ…å« [logits](https://developers.google.com/machine-learning/glossary#logits) æˆ– [log-odds](https://developers.google.com/machine-learning/glossary#log-odds) åˆ†æ•°çš„å‘é‡ï¼Œæ¯ä¸ªç±»ä¸€ä¸ªã€‚

```python
predictions = model(x_train[:1]).numpy()
predictions
```

`tf.nn.softmax` å‡½æ•°å°†è¿™äº› logits è½¬æ¢ä¸ºæ¯ä¸ªç±»çš„*æ¦‚ç‡*ï¼š 

```python
tf.nn.softmax(predictions).numpy()
```

æ³¨ï¼šå¯ä»¥å°† `tf.nn.softmax` çƒ˜ç„™åˆ°ç½‘ç»œæœ€åä¸€å±‚çš„æ¿€æ´»å‡½æ•°ä¸­ã€‚è™½ç„¶è¿™å¯ä»¥ä½¿æ¨¡å‹è¾“å‡ºæ›´æ˜“è§£é‡Šï¼Œä½†ä¸å»ºè®®ä½¿ç”¨è¿™ç§æ–¹å¼ï¼Œå› ä¸ºåœ¨ä½¿ç”¨ softmax è¾“å‡ºæ—¶ä¸å¯èƒ½ä¸ºæ‰€æœ‰æ¨¡å‹æä¾›ç²¾ç¡®ä¸”æ•°å€¼ç¨³å®šçš„æŸå¤±è®¡ç®—ã€‚ 

ä½¿ç”¨ `losses.SparseCategoricalCrossentropy` ä¸ºè®­ç»ƒå®šä¹‰æŸå¤±å‡½æ•°ï¼Œå®ƒä¼šæ¥å— logits å‘é‡å’Œ `True` ç´¢å¼•ï¼Œå¹¶ä¸ºæ¯ä¸ªæ ·æœ¬è¿”å›ä¸€ä¸ªæ ‡é‡æŸå¤±ã€‚

```python
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
```

æ­¤æŸå¤±ç­‰äº true ç±»çš„è´Ÿå¯¹æ•°æ¦‚ç‡ï¼šå¦‚æœæ¨¡å‹ç¡®å®šç±»æ­£ç¡®ï¼Œåˆ™æŸå¤±ä¸ºé›¶ã€‚

è¿™ä¸ªæœªç»è®­ç»ƒçš„æ¨¡å‹ç»™å‡ºçš„æ¦‚ç‡æ¥è¿‘éšæœºï¼ˆæ¯ä¸ªç±»ä¸º 1/10ï¼‰ï¼Œå› æ­¤åˆå§‹æŸå¤±åº”è¯¥æ¥è¿‘ `-tf.math.log(1/10) ~= 2.3`ã€‚

```python
loss_fn(y_train[:1], predictions).numpy()
```

åœ¨å¼€å§‹è®­ç»ƒä¹‹å‰ï¼Œä½¿ç”¨ Keras `Model.compile` é…ç½®å’Œç¼–è¯‘æ¨¡å‹ã€‚å°† [`optimizer`](https://tensorflow.google.cn/api_docs/python/tf/keras/optimizers) ç±»è®¾ç½®ä¸º `adam`ï¼Œå°† `loss` è®¾ç½®ä¸ºæ‚¨ä¹‹å‰å®šä¹‰çš„ `loss_fn` å‡½æ•°ï¼Œå¹¶é€šè¿‡å°† `metrics` å‚æ•°è®¾ç½®ä¸º `accuracy` æ¥æŒ‡å®šè¦ä¸ºæ¨¡å‹è¯„ä¼°çš„æŒ‡æ ‡ã€‚

```python
model.compile(optimizer='adam',
              loss=loss_fn,
              metrics=['accuracy'])
```

#### 2.3 è®­ç»ƒå¹¶è¯„ä¼°æ¨¡å‹ <a id='2-3'></a>

ä½¿ç”¨ `Model.fit` æ–¹æ³•è°ƒæ•´æ‚¨çš„æ¨¡å‹å‚æ•°å¹¶æœ€å°åŒ–æŸå¤±ï¼š 

```python
model.fit(x_train, y_train, epochs=5)
```

`Model.evaluate` æ–¹æ³•é€šå¸¸åœ¨ "[Validation-set](https://developers.google.com/machine-learning/glossary#validation-set)" æˆ– "[Test-set](https://developers.google.com/machine-learning/glossary#test-set)" ä¸Šæ£€æŸ¥æ¨¡å‹æ€§èƒ½ã€‚

```python
model.evaluate(x_test,  y_test, verbose=2)
```

ç°åœ¨ï¼Œè¿™ä¸ªç…§ç‰‡åˆ†ç±»å™¨çš„å‡†ç¡®åº¦å·²ç»è¾¾åˆ° 98%ã€‚æƒ³è¦äº†è§£æ›´å¤šï¼Œè¯·é˜…è¯» [TensorFlow æ•™ç¨‹](https://tensorflow.google.cn/tutorials/)ã€‚

å¦‚æœæ‚¨æƒ³è®©æ¨¡å‹è¿”å›æ¦‚ç‡ï¼Œå¯ä»¥å°è£…ç»è¿‡è®­ç»ƒçš„æ¨¡å‹ï¼Œå¹¶å°† softmax é™„åŠ åˆ°è¯¥æ¨¡å‹ï¼š

```python
probability_model = tf.keras.Sequential([
  model,
  tf.keras.layers.Softmax()
])
```

```python
probability_model(x_test[:5])
```
